```python
##codigo para salvar em PDF o conteúdo de um jupyter notebook

!pip install pdfkit

import subprocess, pdfkit
from pathlib import Path
import ipynbname

nb_path = Path(ipynbname.path())
out_dir = Path(r"D:\VSCODE_OUTPUT")
out_dir.mkdir(parents=True, exist_ok=True)

# 1) Notebook → HTML
html_dir = out_dir / "_tmp_html"
html_dir.mkdir(exist_ok=True)
subprocess.check_call([
    "jupyter", "nbconvert",
    "--to", "html",
    "--no-input",
    str(nb_path),
    "--output-dir", str(html_dir),
    "--output", nb_path.stem
])

html_path = html_dir / f"{nb_path.stem}.html"
pdf_path  = out_dir / f"{nb_path.stem}.pdf"

# 2) HTML → PDF
pdfkit.from_file(str(html_path), str(pdf_path))
print(f"✅ PDF saved as: {pdf_path}")
```
